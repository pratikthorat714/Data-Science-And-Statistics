```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

# Line Chart

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("Line Chart")
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.bar(x,y,color='red')
plt.title("bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Scatter Plot

```python
a=(1,2,3,5,7,8,9)
b=(10,22,44,50,60,63,36)
plt.scatter(a,b)
plt.title("scatter chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Pieplot Chart

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 20, 20]  
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  
plt.title('Fruit Distribution')
plt.show()
```
